```python
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize = (10, 10))
 
# syntax for 3-D projection
Three_D = plt.axes(projection ='3d')
 
# defining axes for Baloon
Baloon_width = np.random.normal(10,2,350) # X axis
Baloon_length = np.random.normal(10,5 ,350) # Y axis
Baloon_weight = np.random.normal(1, 0.3, 350) # Z axis

# defining axes for Melon
Melon_width = np.random.normal(10,2,350) # X axis
Melon_length = np.random.normal(15,3 ,350) # Y axis
Melon_weight = np.random.normal(7, 3, 350) # Z axis


Three_D.scatter(Baloon_width, Baloon_length, Baloon_weight, alpha = 0.5, marker = 'o')
 
Three_D.scatter(Melon_width, Melon_length, Melon_weight, alpha = 0.5, marker = '^')

Three_D.set_xlabel('Width')
Three_D.set_ylabel('Length')
Three_D.set_zlabel('Weight')

plt.show()
```
