```python
from scipy import stats as sts
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = sts.linregress(x,y)

def my_fun(x):
    return slope * x + intercept
    
My_Model = list(map(my_fun, x))

plt.scatter(x, y)
plt.plot(x, My_Model)


plt.show()
```

```python
from scipy import stats as sts
import matplotlib.pyplot as plt

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = sts.linregress(x,y)

Label = "relationship: %0.3f"%(r)

def my_fun(x):
    return slope * x + intercept
    
My_Model = list(map(my_fun, x))

plt.scatter(x, y)
plt.plot(x, My_Model, label=Label)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score


x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]


My_Model = np.poly1d(np.polyfit(x,y, 3))

My_Line = np.linspace(1 ,22 ,100)


plt.rcParams["figure.figsize"] = [7.50, 3.50]

Label = "R_Squared: %2.2f" %(r2_score(y, My_Model(x)))
speed = "Speed prediction at 17:00 is: %2.2f" %(My_Model(17))


plt.plot(My_Line, My_Model(My_Line), color='darkblue', label = Label)
plt.annotate(speed ,(17 ,My_Model(17)),(20,90) ,arrowprops = {})
plt.scatter(x,y, color= 'hotpink')


plt.xlabel("Hour of the day")
plt.ylabel("Speed")
plt.legend(loc='lower right')

plt.show()
```

```python
import pandas as pd
from sklearn import linear_model

df = pd.read_csv("cars.csv")


X = df[['Weight','Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X.values, y)

# Predict the CO2 emission of a car where the weight is 2300, and the Volume is 1300
predictedCO2_1 = regr.predict([[2300, 1300]])
predictedCO2_2 = regr.predict([[3300, 1300]])

print("Predicted CO2 \n(Weight:2300 / Volume: 1300):   {}\n(Weight:3300 / Volume: 1300):   {}\nCoefficient is:    {}".format(predictedCO2_1,predictedCO2_1,regr.coef_))
```

```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

df = pd.read_csv("cars2.csv")

X = df[['Weight' , 'Volume']]
y = df['CO2']


scaled_X = scale.fit_transform(X.values)

regr = linear_model.LinearRegression()
regr.fit(scaled_X, y)

scaled = scale.transform([[2400, 1.5]])

predict_CO2 = regr.predict([scaled[0]])

print("Predicted CO2:    {}".format(predict_CO2))

print("Scaled_X first five values: \n",scaled_X[:5])
```
