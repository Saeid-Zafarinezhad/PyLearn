```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([330,320,310,300,290,280,250])
y = np.array([8,8.5,9,9.5,10,10.5,11])

plt.plot(x, y)

font1 = {'family':'calibri','color':'darkblue','size':15}
font2 = {'family':'serif','color':'darkred','size':10}

plt.title("Fuel Consumption" ,loc='right' ,fontdict=font1)
plt.xlabel("Average speed" ,fontdict=font2)
plt.ylabel("Fuel Burnage", fontdict=font2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([10,25,45,35,20])

M_Labels = ["Services","Insurance","Salary","Expenses", "Vacation"]
M_Explode = [0.15,0,0.09,0,0]

plt.pie(y, labels = M_Labels ,startangle = 45 ,explode = M_Explode, shadow=True)
plt.legend(title="Expenses Survey", loc='best', bbox_to_anchor=(0.9, 0.9, 0.5, 0.5))
plt.show()
```
